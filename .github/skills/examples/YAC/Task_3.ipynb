# Task 3: Run coupled setup with additional Python components

- data sent by a component, can be received by more than one component
- the coupling configuration can be queried by any component
- the following coupled run includes additional Python scripts that extract data and information from the original configuration

```bash
mpirun \
    -n 4 toy_atm.x : \
    -n 4 toy_ocn.x : \
    -n 1 ./python/dump_metadata.py : \
    -n 1 ./python/simple_output.py \
        --res 720 360 sst.nc ocn_comp ocn_grid sea_surface_temperature
```

- `dump_metadata.py` will generate the file [metadata.yaml](metadata.yaml) containing most information that can be queried at runtime through the YAC interface by all processes
- `simple_output.py` will get the 'sea_surface_temperature' field from `toy_ocn.x` and will write it to `sst.nc` on a regular lon lat grid
  - visualise `sst.nc` e.g. using `xarray`

```bash
python -c "
import xarray
xarray.open_dataset('sst.nc')['sea_surface_temperature'].isel(time=3).plot().figure.savefig('sst.png')"
echo "<img src='sst.png' />" | displayHTML
```
