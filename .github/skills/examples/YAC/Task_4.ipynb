# Task 4: Difference between MPMD, SPMD, and MPMD+SPMD
- MPMD vs. SPMD (Multiple/Single Program Multiple Data)
- current coupled toy setup is MPMD
  - each component has its own executable (`toy_atm.x`, `toy_ocn.x`, `dump_metadata.py`, and `simple_output.py`)
- ICON is SPMD
  - has only a single executable (`icon`)
  - after initialisation ICON processes will take specific roles (`atm`, `ocn`, `wave`, `atm-output`, ...)
  - roles are determined by their rank within the global communicator
- ICON in a coupled setup with external components
  - at initialisation ICON has to first generate an ICON-only communicator, before it can determine the roles of its processes and register them accordingly in YAC
- `toy_spmd.x` distributes its processes evenly to the atmosphere and ocean roles
  - **Task**: adjust it, such that it first creates a "spmd"-communicator and then use it to initialise YAC (see [Initialising YAC](https://dkrz-sw.gitlab-pages.dkrz.de/yac/d4/d40/init_yac_detail.html))
