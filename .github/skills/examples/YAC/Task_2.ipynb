## `plot_weights.py` script
YAC provides the [plot_weights.py](YAC/yac/tools/plot_weights.py) tool to visualize the weights file.

- visualise weights over Gibraltar (5°W 36°N)

```bash
python YAC/yac/tools/plot_weights.py \
    icon:grids/icon_grid_0036_R02B04_O.nc \
    icon:grids/icon_grid_0030_R02B03_G.nc \
    sst_weights.nc \
    --save_as sst_weights.png \
    --center -5 36
echo "<img src='sst_weights.png' />" | displayHTML
```
