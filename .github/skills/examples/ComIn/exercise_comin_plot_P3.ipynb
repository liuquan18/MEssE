This is a Python 3 Jupyter notebook. We create a plot Matplotlib and Cartopy and start by loading the necessary modules.

```python
import getpass
import numpy as np
import xarray as xr
import cartopy
import matplotlib.pyplot as plt
```

Define the directory and filenames.

```python
user = getpass.getuser()

# absolute path to directory with plenty of space:
SCRATCHDIR = f"/scratch/{user[0]}"
EXPDIR = f"{SCRATCHDIR}/{user}/icon_exercise_comin/"
# file names
filename_ll = f"{EXPDIR}/NWP_LAM_DOM01_20210714T120000Z.nc"
print(f"Filename:  {filename_ll=}")
```

```python
plotvar1='RHI_MAX'
plotvar2 = 'QI_MAX'
```

### Create the plot

```python
#%matplotlib widget

ds_ll = xr.open_dataset(filename_ll)
lon   = ds_ll[plotvar1].lon
lat   = ds_ll[plotvar1].lat

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(18, 9), subplot_kw={'projection': cartopy.crs.PlateCarree(), 'frameon': False})
plot_color = "viridis"
# Plot RHI_MAX
dataset1 = ds_ll[plotvar1][0, :, :]
plot_min1 = np.round(np.amin(dataset1))
plot_max1 = np.round(np.amax(dataset1))

ax1 = axs[0]
ax1.set_title("RHI_MAX", y=1.05)
ax1.add_feature(cartopy.feature.BORDERS, edgecolor='gray')
cf1 = ax1.contourf(lon, lat, dataset1, cmap=plot_color, transform=cartopy.crs.PlateCarree())
ax1.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)
plt.colorbar(cf1, ax=ax1, orientation='horizontal', label="Maximum relative humidity over ice [%]")

# Plot QI_MAX
dataset2 = ds_ll[plotvar2][0, :, :]
plot_min2 = np.round(np.amin(dataset2))
plot_max2 = np.round(np.amax(dataset2))

ax2 = axs[1]
ax2.set_title("QI_MAX", y=1.05)
ax2.add_feature(cartopy.feature.BORDERS, edgecolor='gray')
cf2 = ax2.contourf(lon, lat, dataset2, cmap=plot_color, transform=cartopy.crs.PlateCarree())
ax2.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)
plt.colorbar(cf2, ax=ax2, orientation='horizontal', label="Maximum cloud ice content [kg/kg]")

# Show the plot
plt.show()
```
